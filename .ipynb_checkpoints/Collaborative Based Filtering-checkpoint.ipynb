### Implementing Collaborative Filtering

<b> Item Based Collaborative Filtering </b> <br>
This is the case to find similar movies instead to similar users. <br>
To find every pair of items that were rated/viewed/clicked by same user, then measuring the similarity of those rated/viewed/liked/clicked across all user who rated/viewed/liked/clicked. Finally recommending them based on similarity scores. <br>
<br>
<b> Example </b> </br>
If John Likes <b> A, B, C movie </b>then,  <br>
we found that <b> D movie </b> is highly similar to <b> C movie</b>. <br>
Hence, There is a chance that John will like <b> D movie </b>

<br>
Item-based Collaborative Filtering is the method to find the items similar to the ones user has already liked.

```python
# Converting to Matrix form
final_dataset = ratings_df.pivot(index='movieId',columns='userId',values='rating')
final_dataset.head()
```

Since, not all the users won't be rating all the movie thus, there are values with NaN /Null values. <br>
These values will be treated by replacing them with 0.


```python
# Normalize user-movie matrix
final_dataset = final_dataset.subtract(final_dataset.mean(axis=1), axis = 0)
final_dataset.head()
```

The movie rating with more than avg get +ve value and less than avg get -ve value

```python
# Filling twhe Null values with 0
final_dataset.fillna(0, inplace = True)
final_dataset.head()
```

<b> Removing Noise </b> <br>
Reducing noise by filtering movies and users

```python
# Removing noise from the data
no_user_voted= ratings_df.groupby('movieId')['rating'].agg('count')
no_movies_voted = ratings_df.groupby('userId')['rating'].agg('count')
```

#### PLotting for Users

```python
plt.figure(figsize= (10, 7))

color_ = no_user_voted.index
plt.scatter(no_user_voted.index,no_user_voted, alpha = 0.5, c = color_)
plt.axhline(y =10, color = 'black') # Base Count of 10
plt.grid('True')
plt.xlabel('MovieId')
plt.ylabel('No. of users voted')
plt.show()
```

```python
# Filtering out the users under 10 count for voted (Removed)
final_dataset = final_dataset.loc[no_user_voted[no_user_voted > 10].index,:]
```

#### Plotting for movies

```python
plt.figure(figsize = (10, 7))
plt.scatter(no_movies_voted.index,no_movies_voted,color='red', alpha = 0.5)
plt.axhline(y=20,color='black')
plt.xlabel('UserId')
plt.ylabel('No. of votes by user')
plt.show()
```

```python
# Filtering out the movies with less than 50 ratings
final_dataset=final_dataset.loc[:, no_movies_voted[no_movies_voted > 50].index]
final_dataset
```

### What is Sparsity Problem?
Data sparsity arises from the phenomenon that users in general rate only limited number of items. Thus, it's not credible enough. <br> <br>

#### Removing Sparsity
Sparse Data is the data that has mostly elements that don't carry any information. <br>
Example, Where most values are zero. Similar to above dataset <br>

```python
from scipy.sparse import csr_matrix

# Compressed Sparse Row
csr_data = csr_matrix(final_dataset.values)
final_dataset.reset_index(inplace=True)

# Its also done to reduce Spatial Complexity
```

```python
# Checking out dataset
final_dataset.head()
```

### Implementation

```python
from sklearn.neighbors import NearestNeighbors

# Since our dataset is small, Hence Brute Force is used
knn = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=20, n_jobs=-1)
knn.fit(csr_data)
```

### Recommending Movies

```python
def get_movie_recommendation(movie_name):
    
    # Number of movies to recommend
    num_movies_reccomend = 10
    
    # Since, similarity is required hence movie name must exist in the list
    # Checking if the movie name exist in the list
    movie = movies_df[movies_df['title'].str.contains(movie_name)]  
    
    if len(movie): 
        # Fetching Movie Id for the given movie name
        movie_idx = movie.iloc[0]['movieId']
        
        # Converting Movie Id to index
        movie_idx = final_dataset[final_dataset['movieId'] == movie_idx].index[0]
        
        # distances and movieId model predicting
        distances , indices = knn.kneighbors(csr_data[movie_idx], n_neighbors = num_movies_reccomend+1)
        
        # Since the values need to be sorted according to distance,Thus using labda function to fetch distance values
        rec_movie_indices = sorted(list(zip(indices.squeeze().tolist(),distances.squeeze().tolist())),key=lambda x: x[1])[::-1]
        
        return rec_movie_indices
    else:
        return "Movie Name Not in the List"
```

```python
# Checking how movie id and indexes looks like
rec_indices = get_movie_recommendation("Superman")
for i in rec_indices:
    print(i)
```

```python
# Printing the recommended movie names
dist = []
for i in rec_indices:
    # fetching value of movieID from final dataset
    movie_index = final_dataset.loc[i[0]]['movieId']
    
    # assigning index values from movies_df using movie_index from final dataset
    idx = movies_df[movies_df['movieId'] == movie_index].index
    
    # printing using index values of movies
    print((movies_df.loc[idx]['title'].values[0]))
    dist.append(i[1])
```
